### 一）可调整的资产负债表指标

#### DSRI（Days' sales in receivable index）：应收账款指数

显然DSRI指数过高，表明公司可能操控应收账款项目来对营业收入与利润进行控制。

```python
####################可调整的资产负债表指标
#DSRI（Days' sales in receivable index）：应收账款指数
#DSRI=(本期应收账款⁄本期营业收入)/(上期应收账款⁄上期营业收入)
#DSRI = (Net Receivablest / Salest) / Net Receivablest-1 / Salest-1)
dsri_element_2011 = df2011[df2011.filter(regex='Unnamed: 0|ACCT_RCV|TOT_OPER_REV').columns]
dsri_element_2012 = df2012[df2012.filter(regex='Unnamed: 0|ACCT_RCV|TOT_OPER_REV').columns]
dsri_element_2013 = df2013[df2013.filter(regex='Unnamed: 0|ACCT_RCV|TOT_OPER_REV').columns]
dsri_element_2014 = df2014[df2014.filter(regex='Unnamed: 0|ACCT_RCV|TOT_OPER_REV').columns]
dsri_element_2015 = df2015[df2015.filter(regex='Unnamed: 0|ACCT_RCV|TOT_OPER_REV').columns]
dsri_element_2016 = df2016[df2016.filter(regex='Unnamed: 0|ACCT_RCV|TOT_OPER_REV').columns]
dsri_element_2017 = df2017[df2017.filter(regex='Unnamed: 0|ACCT_RCV|TOT_OPER_REV').columns]

# combine the above tables
from functools import reduce
dfs = [dsri_element_2011, dsri_element_2012, dsri_element_2013, dsri_element_2014, dsri_element_2015,  dsri_element_2016, dsri_element_2017]
dsri_element = reduce(lambda left,right: pd.merge(left,right,on='Unnamed: 0', how= 'outer'), dfs)
#fill the na at first
dsri_element.fillna('0',inplace = True)
dsri_element
```

```python
#  to create features using for loop
def foo(x,y):
    try:
        return x / y
    except ZeroDivisionError:
        return 0

n = 2
for i in range(4,len(dsri_element.columns),2):
    dsri_element['DSRI_201'+str(n)] = ((dsri_element[dsri_element.columns[i]].apply(lambda x:float(x)))/((dsri_element[dsri_element.columns[i-1]]).apply(lambda x:float(x)))) /((((dsri_element[dsri_element.columns[i-2]]).apply(lambda x:float(x)))/((dsri_element[dsri_element.columns[i-3]]).apply(lambda x:float(x)))).replace((np.inf, -np.inf), (0, 0)))
    n += 1
    

# fill missing value and inf
dsri_element.fillna('0',inplace = True)
dsri_element.replace((np.inf, -np.inf), (0, 0), inplace=True)
DSRI = dsri_element.filter(like = 'DSRI')
DSRI
```

#### AQI（Asset quality index）：资产质量指数

AQI指数检测公司是否可能通过操控非实物资产项目来控制利润率。

```python
#AQI（Asset quality index）：资产质量指数
#AQI=本期非实物资产比例/上期非实物资产比例
#Asset Quality Index = Proportion of Current and Fixed Assets in Period 1 / Proportion of Current and Fixed Assets in Period 1
#Where:
#Proportion of Current and Fixed Assets = 1 - (Current Assets + Net Fixed Assets) / Total Assets

aqi_element_2011 = df2011[df2011.filter(regex='Unnamed: 0|TOT_CUR_ASSETS|FIX_ASSETS|TOT_ASSETS').columns]
aqi_element_2012 = df2012[df2012.filter(regex='Unnamed: 0|TOT_CUR_ASSETS|FIX_ASSETS|TOT_ASSETS').columns]
aqi_element_2013 = df2013[df2013.filter(regex='Unnamed: 0|TOT_CUR_ASSETS|FIX_ASSETS|TOT_ASSETS').columns]
aqi_element_2014 = df2014[df2014.filter(regex='Unnamed: 0|TOT_CUR_ASSETS|FIX_ASSETS|TOT_ASSETS').columns]
aqi_element_2015 = df2015[df2015.filter(regex='Unnamed: 0|TOT_CUR_ASSETS|FIX_ASSETS|TOT_ASSETS').columns]
aqi_element_2016 = df2016[df2016.filter(regex='Unnamed: 0|TOT_CUR_ASSETS|FIX_ASSETS|TOT_ASSETS').columns]
aqi_element_2017 = df2017[df2017.filter(regex='Unnamed: 0|TOT_CUR_ASSETS|FIX_ASSETS|TOT_ASSETS').columns]

# combine the above tables
from functools import reduce
dfs2 = [aqi_element_2011, aqi_element_2012, aqi_element_2013, aqi_element_2014, aqi_element_2015,  aqi_element_2016, aqi_element_2017]
aqi_element = reduce(lambda left,right: pd.merge(left,right,on='Unnamed: 0', how= 'outer'), dfs2)

#fill the na at first
aqi_element.fillna('0',inplace = True)
aqi_element
```

```python
# convert the dtype from object to float to make caluculation free from error
for i in range(1,22):
    aqi_element.iloc[:,i] = aqi_element.iloc[:,i].apply(lambda x:float(x))
```

```python
# to create AQI features using for loop
n = 2
for i in range(6,len(aqi_element.columns),3):
    aqi_element['AQI_201'+str(n)] = (1 - ((aqi_element[aqi_element.columns[i]]+aqi_element[aqi_element.columns[i-2]])/aqi_element[aqi_element.columns[i-1]])) / (1 - ((aqi_element[aqi_element.columns[i-3]]+aqi_element[aqi_element.columns[i-5]])/aqi_element[aqi_element.columns[i-4]]))
    n += 1

# fill missing value and inf
aqi_element.fillna('0',inplace = True)
aqi_element.replace((np.inf, -np.inf), (0, 0), inplace=True)

# EXTRACT AQI columns
AQI = aqi_element.filter(like = 'AQI') 
AQI
```

#### TATA（Total accruals to total assets）：总应计项

TATA指数检测公司的非货币类的流动资产、流动负债有无异常变动情况，因为非货币类或长期应计项的认为操控也可以直接影响到最终的利润结果。

```python
# TATA（Total accruals to total assets）：总应计项
# TATA=应计项/总资产=((∆流动资产-∆货币资金)-(∆流动负债-∆一年内到期长期负债-∆应交税费)-折旧费用)/总资产
#Accruals to Assets Ratio = (∆ Working Capital - ∆ Cash - ∆ Depreciation) / ∆ Total Assets
#Where:
#The change (∆) in working capital, cash and depreciation are measured in terms of the observed difference in these accounts occurring between two accounting periods.
tata_element_2011 = df2011[df2011.filter(regex='Unnamed: 0|TOT_CUR_ASSETS|MONETARY_CAP|TOT_CUR_LIAB|NON_CUR_LIAB_DUE_WITHIN_1Y|TAXES_SURCHARGES_PAYABLE|TOT_ASSETS').columns]
tata_element_2012 = df2012[df2012.filter(regex='Unnamed: 0|TOT_CUR_ASSETS|MONETARY_CAP|TOT_CUR_LIAB|NON_CUR_LIAB_DUE_WITHIN_1Y|TAXES_SURCHARGES_PAYABLE|TOT_ASSETS').columns]
tata_element_2013 = df2013[df2013.filter(regex='Unnamed: 0|TOT_CUR_ASSETS|MONETARY_CAP|TOT_CUR_LIAB|NON_CUR_LIAB_DUE_WITHIN_1Y|TAXES_SURCHARGES_PAYABLE|TOT_ASSETS').columns]
tata_element_2014 = df2014[df2014.filter(regex='Unnamed: 0|TOT_CUR_ASSETS|MONETARY_CAP|TOT_CUR_LIAB|NON_CUR_LIAB_DUE_WITHIN_1Y|TAXES_SURCHARGES_PAYABLE|TOT_ASSETS').columns]
tata_element_2015 = df2015[df2015.filter(regex='Unnamed: 0|TOT_CUR_ASSETS|MONETARY_CAP|TOT_CUR_LIAB|NON_CUR_LIAB_DUE_WITHIN_1Y|TAXES_SURCHARGES_PAYABLE|TOT_ASSETS').columns]
tata_element_2016 = df2016[df2016.filter(regex='Unnamed: 0|TOT_CUR_ASSETS|MONETARY_CAP|TOT_CUR_LIAB|NON_CUR_LIAB_DUE_WITHIN_1Y|TAXES_SURCHARGES_PAYABLE|TOT_ASSETS').columns]
tata_element_2017 = df2017[df2017.filter(regex='Unnamed: 0|TOT_CUR_ASSETS|MONETARY_CAP|TOT_CUR_LIAB|NON_CUR_LIAB_DUE_WITHIN_1Y|TAXES_SURCHARGES_PAYABLE|TOT_ASSETS').columns]
#there is no TAXES_SURCHARGES_PAYABLE

# combine the above tables
from functools import reduce
dfs3 = [tata_element_2011, tata_element_2012, tata_element_2013, tata_element_2014, tata_element_2015,  tata_element_2016, tata_element_2017]
tata_element = reduce(lambda left,right: pd.merge(left,right,on='Unnamed: 0', how= 'outer'), dfs3)

#fill the na at first
tata_element.fillna('0',inplace = True)
tata_element
```

```python
#TATA=应计项/总资产=((∆流动资产-∆货币资金)-(∆流动负债-∆一年内到期长期负债-∆应交税费)-折旧费用)/总资产
n = 2
for i in range(9,len(tata_element.columns),5):
    tata_element['TATA_201'+str(n)] = (((tata_element[tata_element.columns[i]]-tata_element[tata_element.columns[i-5]])-(tata_element[tata_element.columns[i-3]]-tata_element[tata_element.columns[i-8]]))-((tata_element[tata_element.columns[i+1]]-tata_element[tata_element.columns[i-4]])-(tata_element[tata_element.columns[i-2]]-tata_element[tata_element.columns[i-7]])))/(tata_element[tata_element.columns[i-1]]-tata_element[tata_element.columns[i-6]])
    n += 1

# fill missing value and inf
tata_element.fillna('0',inplace = True)
tata_element.replace((np.inf, -np.inf), (0, 0), inplace=True)

#extract TATA cols
TATA = tata_element.filter(like = 'TATA') 
TATA
```

### 三）可调整的收入类指标

#### GMI（Gross margin index）：毛利率指数

GMI指数检测毛利率是否有异常变化。当毛利率指数大于1时，意味着毛利率缩小了。原公式认为毛利率缩小是一个消极的信号，前景不明朗的企业更有可能操纵盈余。

```python
#GMI（Gross margin index）：毛利率指数
#计算公式：GMI=上期毛利率/本期毛利率
#GMI = [(Salest-1 - COGSt-1) / Salest-1] / [(Salest - COGSt) / Salest]
gmi_element_2011 = df2011[df2011.filter(regex='Unnamed: 0|OPER_REV|LESS_OPER_COST').columns]
gmi_element_2012 = df2012[df2012.filter(regex='Unnamed: 0|OPER_REV|LESS_OPER_COST').columns]
gmi_element_2013 = df2013[df2013.filter(regex='Unnamed: 0|OPER_REV|LESS_OPER_COST').columns]
gmi_element_2014 = df2014[df2014.filter(regex='Unnamed: 0|OPER_REV|LESS_OPER_COST').columns]
gmi_element_2015 = df2015[df2015.filter(regex='Unnamed: 0|OPER_REV|LESS_OPER_COST').columns]
gmi_element_2016 = df2016[df2016.filter(regex='Unnamed: 0|OPER_REV|LESS_OPER_COST').columns]
gmi_element_2017 = df2017[df2017.filter(regex='Unnamed: 0|OPER_REV|LESS_OPER_COST').columns]

# combine the above tables
from functools import reduce
dfs5 = [gmi_element_2011, gmi_element_2012, gmi_element_2013, gmi_element_2014, gmi_element_2015,  gmi_element_2016, gmi_element_2017]
gmi_element = reduce(lambda left,right: pd.merge(left,right,on='Unnamed: 0', how= 'outer'), dfs5)

#fill the na at first
gmi_element.fillna('0',inplace = True)
sgi_element = gmi_element.copy() 
gmi_element
```

```python
# convert the dtype from object to float to make calculation free from error
for i in range(1,22):
    gmi_element.iloc[:,i] = gmi_element.iloc[:,i].apply(lambda x:float(x))
```

```python
#GMI = [(Salest-1 - COGSt-1) / Salest-1] / [(Salest - COGSt) / Salest]
n = 2
for i in range(6,len(gmi_element.columns),3):
    gmi_element['GMI_201'+str(n)] = ((gmi_element[gmi_element.columns[i-3]]- gmi_element[gmi_element.columns[i-5]])/gmi_element[gmi_element.columns[i-3]])/((sgai_element[sgai_element.columns[i]]-sgai_element[sgai_element.columns[i-2]])/sgai_element[sgai_element.columns[i]])
    n += 1

    
# fill missing value and inf
gmi_element.fillna('0',inplace = True)
gmi_element.replace((np.inf, -np.inf), (0, 0), inplace=True)

#extract GMI cols
GMI = gmi_element.filter(like = 'GMI') 
GMI
```

```python
#convert the data type from object to float again
n=2
for i in range(6):
    GMI['GMI_201'+str(n)] = GMI['GMI_201'+str(n)].apply(lambda x:float(x))
    n +=1

#check
GMI.info()
```

#### SGI（Sales growth index）：营业收入指数

SGI指数检验营业收入是否有异常变动，操控利润通常都要通过操控营业收入。销售增长指数指第本期的销售额与上期的销售额的比率。如果比例增长并不意味着操纵收益，但因为增长中的公司更加容易有财务报表操纵的动力和压力。

```python
# convert the dtype from object to float to make calculation free from error
for i in range(1,22):
    sgi_element.iloc[:,i] = sgi_element.iloc[:,i].apply(lambda x:float(x))
```

#### LVGI（Leverage index）：财务杠杆指数

LVGI指数检测公司的债务状况有无变动。原公式认为如果本期负债率上升，财务操纵的可能性是降低的，因为发展中公司偏向于借债，偿还能力佳，并且将负债显示在报表上更加增加财务可信度。

```python
#计算公式：LVGI=本期资产负债率/上期资产负债率
#LVGI = [(Current Liabilitiest + Total Long Term Debtt) / Total Assetst] / [(Current Liabilitiest-1 + Total Long Term Debtt-1) / Total Assetst-1]
lvgi_element_2011 = df2011[df2011.filter(regex='Unnamed: 0|TOT_CUR_LIAB|LT_BORROW|TOT_ASSETS').columns]
lvgi_element_2012 = df2012[df2012.filter(regex='Unnamed: 0|TOT_CUR_LIAB|LT_BORROW|TOT_ASSETS').columns]
lvgi_element_2013 = df2013[df2013.filter(regex='Unnamed: 0|TOT_CUR_LIAB|LT_BORROW|TOT_ASSETS').columns]
lvgi_element_2014 = df2014[df2014.filter(regex='Unnamed: 0|TOT_CUR_LIAB|LT_BORROW|TOT_ASSETS').columns]
lvgi_element_2015 = df2015[df2015.filter(regex='Unnamed: 0|TOT_CUR_LIAB|LT_BORROW|TOT_ASSETS').columns]
lvgi_element_2016 = df2016[df2016.filter(regex='Unnamed: 0|TOT_CUR_LIAB|LT_BORROW|TOT_ASSETS').columns]
lvgi_element_2017 = df2017[df2017.filter(regex='Unnamed: 0|TOT_CUR_LIAB|LT_BORROW|TOT_ASSETS').columns]

# combine the above tables
from functools import reduce
dfs6 = [lvgi_element_2011, lvgi_element_2012, lvgi_element_2013, lvgi_element_2014, lvgi_element_2015,  lvgi_element_2016, lvgi_element_2017]
lvgi_element = reduce(lambda left,right: pd.merge(left,right,on='Unnamed: 0', how= 'outer'), dfs6)

#fill the na at first
lvgi_element.fillna('0',inplace = True)
lvgi_element
```

```python
# convert the dtype from object to float to make calculation free from error
for i in range(1,22):
    lvgi_element.iloc[:,i] = lvgi_element.iloc[:,i].apply(lambda x:float(x))
```

```python
import pandas as pd

data = pd.read_csv('/Users/jiayangdu/Desktop/PWC/Rating Analysis/ratings.csv')
goodratings = ['AAA','Aa1','Aa2','AAA-','AA+','AA']
data = data.dropna(how='any')

# data['B_INFO_CREDITRATING2'] = [1 if x in goodratings else 0 for x in  data['B_INFO_CREDITRATING']]

data['B_INFO_CREDITRATING2'] = data['B_INFO_CREDITRATING'].apply(lambda x:
    1 if x in goodratings else 0)

data['B_INFO_PRECREDITRATING2'] = data['B_INFO_PRECREDITRATING'].apply(lambda x:
    1 if x in goodratings else 0)

data['label'] = data['B_INFO_PRECREDITRATING2'] - data['B_INFO_CREDITRATING2']
```

```python

# data.loc[data['label'] == 1].to_csv('test.csv', encoding='GBK')

data_downgrade = data.loc[data['label'] == 1]
# data_downgrade.to_csv('data_downgrade.csv', encoding='GBK')

def func(company,date): # data里面 [company] 在这个日期 [之后一年]评级是否下调，下调返回1
    # date:2013 年报的年份
    x = data_downgrade.loc[
        data_downgrade['S_INFO_COMPCODE'] == company]
    if len(x)>0:
        # a:20130601 -> 2013 -> 2012年报信息
        x = x.loc[x['ANN_DT'].apply(lambda a: True if date+1 >= a//10000 > date else False)]
        if len(x) > 0:
            return 1
        return 0
    else:
        return 0

def func1(company,date): # data 里面选出[company]在这个日期[之前][最近的]评级是否为good，好公司返回true
    # data : 2012
    # ANN_DT: 20131201
    x = data.loc[data['S_INFO_COMPCODE']==company]
    if len(x)>0:
        x = x.loc[x['ANN_DT'].apply(lambda a: True if a//10000 <= date else False)]
        if len(x) > 0:
            x = x.sort_values(by='ANN_DT')
            return x.iloc[-1]['B_INFO_CREDITRATING'] in goodratings
        else:
            return False
    else:
        return False

```

```python
# 示例数据
features_all = pd.DataFrame()
for year in [2012,2013,2014,2015,2016,2017]:
    #tempdata = Beneish_Model.copy()
    tempdata = pd.read_csv('/Users/jiayangdu/Desktop/PWC/Rating Analysis/report_{}1231.csv'.format(year), index_col=0)
    tempdata['date'] = year
    # ...
    # tempdata['feature1'] = ...
    # ...
    features_all = pd.concat([features_all,tempdata], axis=0)


# 选出之前评级在2A以上的公司
print('step1')
features_all = features_all.loc[
    features_all.apply(lambda row: func1(row.name,row['date']), axis=1)]

# 把之前评级在2A以上，后续在2A以下的公司打1
print('step2')
features_all['label'] = features_all.apply(lambda row: func(row.name,row['date']), axis=1)

print('writing')
features_all.to_csv('data_for_modeling.csv')
```
